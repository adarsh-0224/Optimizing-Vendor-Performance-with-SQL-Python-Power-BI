```python
import pandas as pd

# Load the 3 cleaned CSV files
vendor_df = pd.read_csv("C:/Users/admin/OneDrive/Desktop/Project 3/vendor_master_clean.csv", sep=";")
item_df = pd.read_csv("C:/Users/admin/OneDrive/Desktop/Project 3/item_master_clean.csv", sep=";")
orders_df = pd.read_csv("C:/Users/admin/OneDrive/Desktop/Project 3/vendor_purchase_orders_clean.csv", sep=";")
```

```python
# Quick Overview
print("\nVendor Master Summary:\n", vendor_df.describe(include='all'))
print("\nItem Master Summary:\n", item_df.describe(include='all'))
print("\nOrders Summary:\n", orders_df.describe(include='all'))
```

# Vendor Performance Analysis

```python
# 1. Total spend per vendor
orders_df["spend"] = orders_df["qty_received"] * orders_df["unit_cost"]
vendor_spend = orders_df.groupby("vendor_name")["spend"].sum().sort_values(ascending=False)

print("\nTop Vendors by Spend:\n", vendor_spend.head(10))

# Plot top 10 vendors by spend
plt.figure(figsize=(10,5))
vendor_spend.head(10).plot(kind="bar", color="skyblue")
plt.title("Top 10 Vendors by Spend")
plt.ylabel("Total Spend")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Vendor fill rate performance
orders_df["fill_rate_pct"] = (orders_df["qty_received"] / orders_df["qty_ordered"]) * 100
print("\nFill Rate Stats:\n", orders_df["fill_rate_pct"].describe())

# Average fill rate by vendor
vendor_fill = orders_df.groupby("vendor_name")["fill_rate_pct"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
vendor_fill.head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Vendors by Fill Rate %")
plt.ylabel("Average Fill Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. On-time delivery rate
on_time_rate = orders_df.groupby("vendor_name")["on_time"].mean().sort_values(ascending=False)
print("\nOn-time Delivery Rates (by vendor):\n", on_time_rate.head(10))


# Item & Category Analysis


# Average unit cost by category
avg_cost_category = orders_df.groupby("category")["unit_cost"].mean().sort_values(ascending=False)
print("\nAverage Cost per Category:\n", avg_cost_category)

plt.figure(figsize=(10,5))
avg_cost_category.plot(kind="bar", color="green")
plt.title("Average Unit Cost by Category")
plt.ylabel("Avg Unit Cost")
plt.xticks(rotation=45)
plt.show()

# Top items by spend
item_spend = orders_df.groupby("item_name")["spend"].sum().sort_values(ascending=False)
print("\nTop 10 Items by Spend:\n", item_spend.head(10))

plt.figure(figsize=(10,5))
item_spend.head(10).plot(kind="bar", color="purple")
plt.title("Top 10 Items by Spend")
plt.ylabel("Total Spend")
plt.xticks(rotation=45)
plt.show()


# Correlation Analysis

numeric_cols = ["qty_ordered", "qty_received", "unit_cost", "lead_time_days", "return_qty", "spend"]
corr = orders_df[numeric_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
